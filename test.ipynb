```python

#new list
i = 0
text_pattern = ""
# the range of pixels (hight and width in the size image)
for y in range(pattern.size[1]):
    for x in range(pattern.size[0]):
        r = pattern.getpixel((x, y))[0]
        g = pattern.getpixel((x, y))[1]
        b = pattern.getpixel((x, y))[2]
        if (r <255 or g <255 or b <255):
            text_pattern += "<span style=\"color: rgb({r},{g},{b})\">".format(r = r,g = g, b = b) + text[i] + "</span>"
            i += 1
            # if the list is done
            if (i >= len(text)):
                #go to the first item
                i = 0        
        else:
            text_pattern += " "
    text_pattern += "<br/>"
```

```python
from IPython.display import display, HTML
def show_html(src):
    return display(HTML(src), metadata=dict(isolated=True))
```

```python
html_tmpl = """<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="utf-8">
    <title>{title}</title>
    <style>
    html {{ min-height: 32em; overflow: hidden; }}
    </style>
</head>
<body>
<div style = "font-family: monospace; font-size: 15px; color: red; white-space: pre-wrap;position: absolute; left: 0%; top: 0%; direction: rtl;"> 
{content}
</div>
</body>
</html>"""
```

```python
interp_src = html_tmpl.format(title="My first HTML page", content=text_pattern)
print(interp_src)
```

```python
show_html(interp_src)
```

```python
def mkdiv(content, **kwargs):
    if 'position' not in kwargs:
        kwargs['position'] = 'absolute'
    style_str = ' '.join([": ".join((k.replace('_', '-'), v))+";" for k, v in kwargs.items()])
    return f"<div style='{style_str}'>{content}</div>"
```

```python
src = ['Grandmas chicken soup','Ingredients:','Chicken','300','gram = pound?','6. Carrots','Onions','parsley, root','2', 'liters', 'of water','Cook']
location = [(970, 1100), (960, 1120), (900, 1005), (920, 990), (980, 1100),(1100, 919), (900, 977), (985, 1222), (975, 1320), (910, 970), (990, 960), (975, 970), (1200, 970)]
divs = []
for i in range(len(src)):
    this_div = mkdiv(src[i],
                     top=f"{location[i][1]}px",
                     left=f"{location[i][0]}px",
                     font_size="16pt")
    divs.append(this_div)
interp_src = html_tmpl.format(title="My first HTML page", content=text_pattern +''.join(divs))
print(interp_src)
```
